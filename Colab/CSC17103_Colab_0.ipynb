**Đồ thị là gì?**

Đồ thị (graph) là một cấu trúc toán học bao gồm hai tập hợp: đỉnh (vertices hoặc nodes) và cạnh (edges hoặc links) nối giữa các đỉnh. Đồ thị được biểu diễn bằng G = (V, E) trong đó:
- \(V\) là tập hợp các đỉnh.
- \(E\) là tập hợp các cạnh, mỗi cạnh nối một cặp đỉnh.

**Lý thuyết đồ thị bắt đầu từ đâu?**

Lý thuyết đồ thị bắt đầu từ bài toán Bảy cây cầu Königsberg của Leonhard Euler vào năm 1736. Euler đã chứng minh rằng không thể đi qua tất cả bảy cây cầu một lần mà không quay lại điểm xuất phát. Đây được xem là vấn đề đầu tiên và là sự khởi đầu của lý thuyết đồ thị.

**Có những cách nào để biểu diễn đồ thị?**

Có nhiều cách để biểu diễn đồ thị, bao gồm:

1. **Ma trận kề (Adjacency Matrix)**: Một ma trận vuông, trong đó mỗi phần tử \(A[i][j]\) biểu diễn sự tồn tại (và có thể là trọng số) của cạnh giữa đỉnh \(i\) và đỉnh \(j\).
   
2. **Danh sách kề (Adjacency List)**: Mỗi đỉnh sẽ có một danh sách liệt kê các đỉnh kề mà nó liên kết.

3. **Ma trận sự cố (Incidence Matrix)**: Một ma trận với hàng là các đỉnh và cột là các cạnh. Phần tử \(I[i][j]\) biểu diễn sự kết nối của đỉnh \(i\) với cạnh \(j\).

**Bạn hãy phân biệt đồ thị (graphs) và mạng (networks)?**

Đồ thị (graphs) và mạng (networks) về bản chất là tương tự, đều bao gồm các đỉnh và cạnh. Tuy nhiên, có một số khác biệt chính:

- **Graphs**: Thường được sử dụng trong lý thuyết toán học và khoa học máy tính để nghiên cứu các tính chất hình thức và cấu trúc của các đối tượng.
  
- **Networks**: Thường được sử dụng trong các ứng dụng thực tiễn như mạng máy tính, mạng xã hội, và các hệ thống phân phối. Networks thường bao gồm thêm các thuộc tính và thông tin phụ khác như trọng số cạnh, loại đỉnh, hoặc dữ liệu gắn liền với các đỉnh và cạnh.

**Đồ thị có ứng dụng (có ý nghĩa) gì đối với cuộc sống của bạn và chúng ta?**

Đồ thị có rất nhiều ứng dụng quan trọng trong nhiều lĩnh vực khác nhau:

1. **Khoa học máy tính**: Sử dụng trong thuật toán tìm đường ngắn nhất, lập lịch trình, phân tích mạng và thiết kế mạch.
   
2. **Mạng xã hội**: Phân tích cấu trúc và động thái của mạng lưới người dùng, xác định ảnh hưởng và sự lan truyền thông tin.

3. **Sinh học**: Sử dụng để phân tích mạng lưới protein, gen và hệ sinh thái.

4. **Giao thông**: Tối ưu hóa tuyến đường, quản lý giao thông và logistics.

5. **Internet và mạng máy tính**: Thiết kế và phân tích mạng lưới các máy chủ, tối ưu hóa việc truyền tải dữ liệu.

6. **Kinh tế và tài chính**: Mô hình hóa mối quan hệ giữa các thực thể kinh tế, phân tích thị trường và rủi ro.

Như vậy, lý thuyết đồ thị không chỉ là một lĩnh vực nghiên cứu lý thuyết mà còn có nhiều ứng dụng thực tiễn, giúp chúng ta giải quyết nhiều vấn đề phức tạp trong cuộc sống và các ngành khoa học.

### Đồ thị

Dựa trên những kiến thức về đồ thị mà bạn vừa trình bày ở phần trên, bây giờ chúng ta sẽ thử sử dụng NetworkX để khám phá bằng máy tính.

#### Khởi tạo

````python
# Khởi tạo một đồ thị vô hướng G
G = nx.Graph()
print(G.is_directed()) # Chắc chắn là không rồi, nhưng mà in ra cho biết :-)
````

````python
# Khởi tạo một đồ thị hữu hướng G
H = nx.DiGraph()
print(H.is_directed())  # Chắc chắn là có rồi, nhưng mà in ra cho biết :-)
````

````python
# Thêm một thuộc tính ở mức đồ thị
# Cụ thể, ta sẽ đặt cho nó một cái tên. Ta sẽ đặt tên cho G là Foo, còn H là Bar
G.graph["Name"] = "Foo"
H.graph["Name"] = "Bar"
G.graph, H.graph
````

#### Đỉnh

Có thể bạn đã biết, đỉnh (nút, node) là một thành phần trong đồ thị. Nó có thể có thuộc tính hoặc không có thuộc tính.

````python
# Thêm một đỉnh với các thuộc tính ở mức đỉnh
G.add_node(0, feature=5, label=0)

# Lấy ra xem thuộc tính của đỉnh 0 vừa thêm vào đồ thị vô hướng
node_0_attr = G.nodes[0]
print("Node 0 has the attributes {}".format(node_0_attr))
````

````python
# Thêm nhiều đỉnh với các thuộc tính ở mức đỉnh
G.add_nodes_from([
  (1, {"feature": 1, "label": 1}),
  (2, {"feature": 2, "label": 2})
]) #(node, attrdict)

# Duyệt qua từng đỉnh trong đồ thị
# Note: Khi bạn set data=True thì nó sẽ trả về các thuộc tính của đỉnh
for node in G.nodes(data=True):
  print(node)

# Lấy ra số lượng đỉnh
num_nodes = G.number_of_nodes()
print("G has {} nodes".format(num_nodes)) # Đoán xem có phải là 3 hay không?
````

#### Cạnh

Tương tự như đỉnh, cạnh (edge) cũng là một thành phần bên trong đồ thị. Và nó cũng có thể mang theo các thuộc tính hoặc không.

````python
# Thêm một cạnh nối giữa đỉnh 0 và 1 có trọng số là 0.5
G.add_edge(0, 1, weight=0.5)

# Lấy ra thông tin thuộc tính (cụ thể là trọng số) của cạnh (0, 1)
edge_0_1_attr = G.edges[(0, 1)]
print("Edge (0, 1) has the attributes {}".format(edge_0_1_attr))
````

````python
# Thêm nhiều cạnh với thông tin trọng số cạnh tương ứng vào đồ thị
G.add_edges_from([
  (1, 2, {"weight": 0.3}),
  (2, 0, {"weight": 0.1})
])

# Duyệt qua tất cả các cạnh
# Note: nếu không có tham số data=True, thì chỉ có cạnh sẽ được trả ra.
for edge in G.edges():
  print(edge)

# Lấy ra số lượng cạnh
num_edges = G.number_of_edges()
print("G has {} edges".format(num_edges)) # Đoán xem có phải là 3 hay không?
````

#### Trực quan đồ thị

NetworkX cũng cho phép chúng ta vẽ đồ thị để có thể ngắm nhìn hình dáng của nó :)

````python
# Draw the graph
nx.draw(G, with_labels = True)
````

### Khám phá đồ thị

NetworkX cũng cho phép chúng ta khám phá nhiều thứ. Cụ thể, ta xem xét đỉnh 0 và lân cận của nó.

````python
# Id của đỉnh cần xem xét
node_id = 0

# Tính bậc của đỉnh cần xem xét
print("Node {} has degree {}".format(node_id, G.degree[node_id]))
````

````python
# Lấy ra các lân cận (hàng xóm) của đỉnh cần xem xét
for neighbor in G.neighbors(node_id):
  print("Node {} has neighbor {}".format(node_id, neighbor))
````

Ngoài ra, thư viện này cũng có nhiều phương thức hữu ích khác. Xem xét về việc sử dụng thuật toán PageRank dưới đây

````python
# Số lượng đỉnh
num_nodes = 4

# Khởi tạo một đường đi giống đồ thị, và sau đó thay đổi nó thành đồ thị có hướng
G = nx.DiGraph(nx.path_graph(num_nodes))

# Rồi vẽ nó ra
nx.draw(G, with_labels = True)

# Tính PageRank
pr = nx.pagerank(G, alpha=0.8)
pr
````

````python
dataset = nx.karate_club_graph()

node_degree_lst = list(dataset.degree())

avg_degree = np.mean(np.array(node_degree_lst)[:,1])
med_degree = np.median(np.array(node_degree_lst)[:,1])

max_degree = max(np.array(node_degree_lst)[:,1])
min_degree = np.min(np.array(node_degree_lst)[:,1])

print(f'Dataset: {dataset}:')
print('======================')
print(f'Number of nodes: {dataset.number_of_nodes()}')
print(f'Number of edges: {dataset.number_of_edges()}')
print(f'Maximum degree : {max_degree}')
print(f'Minimum degree : {min_degree}')
print(f'Average degree : {avg_degree}')
print(f'Median degree : {med_degree}')
print('======================')
````

### Định nghĩa một số hàm phụ trợ

````python
def visualize(h, color, epoch=None, loss=None, accuracy=None):
    plt.figure(figsize=(7,7))
    plt.xticks([])
    plt.yticks([])

    if torch.is_tensor(h):
        h = h.detach().cpu().numpy()
        plt.scatter(h[:, 0], h[:, 1], s=140, c=color, cmap="Set2")
        if epoch is not None and loss is not None and accuracy['train'] is not None and accuracy['val'] is not None:
            plt.xlabel((f'Epoch: {epoch}, Loss: {loss.item():.4f} \n'
                       f'Training Accuracy: {accuracy["train"]*100:.2f}% \n'
                       f' Validation Accuracy: {accuracy["val"]*100:.2f}%'),
                       fontsize=16)
    else:
        nx.draw_networkx(h, pos=nx.spring_layout(h, seed=42), with_labels=False,
                         node_color=color, cmap="Set2")
    plt.show()
````

### Cài đặt mạng Neural Đồ thị

Sau khi tìm hiểu về cách tổ chức dữ liệu của PyG, chúng ta sẽ thử cài đặt một mạng GNN đơn giản bằng cách sử dụng **GCN layer** ([Kipf et al. (2017)](https://arxiv.org/abs/1609.02907)).

PyG cài đặt layer này thông qua [`GCNConv`](https://pytorch-geometric.readthedocs.io/en/latest/modules/nn.html#torch_geometric.nn.conv.GCNConv) mà có thể được sử dụng bằng cách đưa vào đặc trưng đỉnh `x` và biểu diễn liên thông đồ thị COO `edge_index`.

**Đầu vào của GNN là gì?** Một GNN nhận đầu vào là một đồ thị $G = (\mathbb{V}, \mathbb{E})$ mà trong đó mỗi đỉnh  $v_i \in \mathbb{V}$ có một vector đặc trưng đầu vào  $X_i^{(0)}$

**Mục tiêu của việc học là gì?** Ta muốn học một hàm $f \to \mathbb{V} \times \mathbb{R}^d$ mà nhận đầu vào là một đỉnh và đặc trưng của nó. Và hàm này trả ra cho ta một _embedding_, tức là một vector mà biểu diễn đỉnh theo một cách phù hợp cho downstream task của chúng ta. Tùy theo từng downstream task mà ta sẽ có cách định nghĩa "phù hợp" ở đây là như nào. Khi ta có các đỉnh và thuộc tính khởi tạo của chúng được ánh xạ về các embedding đã được học của chúng, ta có thể sử dụng các embedding này để dùng cho các tác vụ hồi quy (regression)/ phân lớp (classification) ở các cấp khác nhau bao gồm cấp đỉnh (node-level), cấp cạnh (edge-level), hoặc cấp đồ thị (graph-level).

Trong phần này, chúng ta sẽ sử dụng các embedding đã được học để phân lớp các đỉnh về đúng cộng đồng của nó.

Ta sẽ bắt đầu cài đặt một kiến trúc GNN như sau:


````python
class GCN(torch.nn.Module):
    def __init__(self):
        super().__init__()
        torch.manual_seed(2024)
        self.conv1 = GCNConv(dataset.num_features, 4)
        self.conv2 = GCNConv(4, 4)
        self.conv3 = GCNConv(4, 2)
        self.classifier = Linear(2, dataset.num_classes)

    def forward(self, x, edge_index):
        # Tầng 1
        h = self.conv1(x, edge_index)
        h = h.tanh() # Kích hoạt

        # Tầng 2
        h = self.conv2(h, edge_index)
        h = h.tanh() # Kích hoạt

        # Tầng 3
        h = self.conv3(h, edge_index)
        h = h.tanh() # Kích hoạt

        # Tầng phân lớp
        out = self.classifier(h)

        return out, h
````

Ở đây, đầu tiên ta khởi tạo tất cả các tầng trong `__init__` và định nghĩa computation flow của kiến trúc mạng trong `forward`.

Theo kiến trúc này, ta định nghĩa và chồng **ba tầng tích chập** lên nhau. Mỗi tầng tương ứng với việc tổng hợp thông tin lân cận một bước (1-hop neighborhood) có hướng của mỗi đỉnh, nhưng khi ta kết hợp các tầng này lại với nhau, ta có thể tổng hợp thông tin lân cận ba bước (3-hop neighborhood) của mỗi đỉnh.

Hơn nữa, các tầng `GCNConv` giảm chiều đặc trưng đỉnh về  $2$, tức là $34 \rightarrow 4 \rightarrow 4 \rightarrow 2$. Mỗi tầng `GCNConv` được tăng cường bởi một hàm kích hoạt phi tuyến, và cụ thể ở đây ta sẽ sử dụng hàm [tanh](https://pytorch.org/docs/stable/generated/torch.nn.Tanh.html?highlight=tanh#torch.nn.Tanh). Sau cùng, ta sử dụng một biến đổi tuyến tính ([`torch.nn.Linear`](https://pytorch.org/docs/stable/generated/torch.nn.Linear.html?highlight=linear#torch.nn.Linear)) mà đóng vai trò như một bộ phân lớp để ánh xạ các đỉnh của chúng ta về 1 trong số 4 lớp/ cộng đồng. Ta trả về cả đầu ra của bộ phân lớp cuối cùng cũng như các embedding của đỉnh.

Giờ ta sẽ thử khởi tạo kiến trúc thông việc gọi `GCN()`, in kiến trúc mô hình ra xem, và thử trực quan embedding:)

````python
model = GCN()

_, h = model(data.x, data.edge_index)
print(f'Embedding shape: {list(h.shape)}')

visualize(h, color=data.y)
````

Ta có một nhận xét nhỏ ở đây như sau: trước khi huấn luyện, mô hình này đã tạo ra các đỉnh nhúng gần giống với cấu trúc cộng đồng của đồ thị. Cụ thể, các đỉnh cùng màu (cộng đồng) đã được nhóm chặt chẽ với nhau trong không gian nhúng, mặc dù trọng số của mô hình của chúng ta được khởi tạo **hoàn toàn ngẫu nhiên** và cho đến nay chúng ta chưa hề thực hiện bất kỳ thủ tục huấn luyện nào cả!

Điều này dẫn đến một kết luận rằng GNN đưa vào một thiên vị quy nạp (inductive bias) mạnh, dẫn đến việc các nhúng tương đồng cho các đỉnh gần nhau trong đồ thị đầu vào!.

### Huấn luyện mô hình

Nhưng ta có thể khiến nó tốt hơn không? Hãy thử sử dụng một chiến lược học bán giám sát (semi-supervised learning) hay học diễn dịch (tranductive learning): ta đơn giản huấn luyện một đỉnh trên mỗi lớp nhưng sử dụng toàn bộ dữ liệu đồ thị.

Giống như việc huấn luyện các mô hình Pytorch khác, ta cũng sẽ cần định nghĩa một hàm mất mát, ở đây ta sử dụng [`CrossEntropyLoss`](https://pytorch.org/docs/stable/generated/torch.nn.CrossEntropyLoss.html). Và ta cũng cần định nghĩa một bộ tối ưu gradient, ở đây ta sử dụng [`Adam`](https://pytorch.org/docs/stable/optim.html?highlight=adam#torch.optim.Adam). Sau đó, ta thực hiện nhiều vòng lặp tối ưu hóa, trong đó tại mỗi vòng lặp bao gồm một bước forward (lan truyền tiến) và backward (lan truyền ngược) để tính toán gradient của các tham số mô hình mà tương ứng đến sự mất mát bắt nguồn từ pha lan truyền tiến.

Lưu ý rằng, quá trình học bán giám sát này đạt được bằng
```
loss = criterion(out[data.train_mask], data.y[data.train_mask])
```
Khi ta tính node embedding cho tất cả các đỉnh, ta chỉ sử dụng các đỉnh huấn luyện để tính toán giá trị hàm mất mát. Và trong cài đặt, ta sử dụng `data.train_mask` để phục vụ cho việc lọc ra các dỉnh đó.

Và giờ, ta sẽ thử huấn luyện mô hình GNN nhé :-)

````python
# Kiểm tra xem có gpu không :-)
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
device
````

````python
# Định nghĩa hàm huấn luyện
def train(data):
    # Thiết lập chế độ huấn luyện mô hình
    model.train()

    # Move data to the GPU
    data = data.to(device)

    # Free gradients
    optimizer.zero_grad()

    # Thực hiện lan truyền tiến
    out, h = model(data.x, data.edge_index)

    # Tính toán giá trị hàm mất mát dựa trên các đỉnh huấn luyện
    loss = criterion(out[data.train_mask], data.y[data.train_mask])

    # Tính toán gradient
    loss.backward()

    # Cập nhật tham số dựa trên gradient
    optimizer.step()

    accuracy = {}
    # Tính toán độ chính xác trên tập huấn luyện
    predicted_classes = torch.argmax(out[data.train_mask], dim=1)
    target_classes = data.y[data.train_mask]
    accuracy['train'] = torch.mean((predicted_classes == target_classes).float())

    # Tính toán độ chính xác trên tập validation
    predicted_classes = torch.argmax(out, dim=1)
    target_classes = data.y
    accuracy['val'] = torch.mean((predicted_classes == target_classes).float())

    return loss, h, accuracy
````

````python
# Một số siêu tham số
## Tốc độ học
lr = 0.01
## Số lượng epoch tối đa
max_epoch = 500

# Khởi tạo model và chuyển nó sang device
model = GCN().to(device)

# Định nghĩa hàm mất mát
criterion = torch.nn.CrossEntropyLoss()

# Định nghĩa bộ tối ưu
optimizer = torch.optim.Adam(model.parameters(), lr=lr)
````

````python
# Thực hiện và xem quá trình huấn luyện
for epoch in range(max_epoch):
    loss, h, accuracy = train(data)
    # Thử trực quan embedding cứ mỗi 100 epoch
    if (epoch + 1) % 100 == 0:
        visualize(h, color=data.y.cpu(), epoch=epoch+1, loss=loss, accuracy=accuracy)
        time.sleep(0.3)
````

````python
# YOUR CODE HERE
# Tải dataset
dataset = KarateClub()
data = dataset[0]

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

# Định nghĩa mô hình GCN với 5 tầng GCNConv
class GCN(torch.nn.Module):
    def __init__(self):
        super().__init__()
        torch.manual_seed(2024)
        self.conv1 = GCNConv(dataset.num_features, 8)
        self.conv2 = GCNConv(8, 8)
        self.conv3 = GCNConv(8, 8)
        self.conv4 = GCNConv(8, 4)
        self.conv5 = GCNConv(4, 2)
        self.classifier = Linear(2, dataset.num_classes)

    def forward(self, x, edge_index):
        # Tầng 1
        h = self.conv1(x, edge_index)
        h = h.tanh()  # Kích hoạt

        # Tầng 2
        h = self.conv2(h, edge_index)
        h = h.tanh()  # Kích hoạt

        # Tầng 3
        h = self.conv3(h, edge_index)
        h = h.tanh()  # Kích hoạt

        # Tầng 4
        h = self.conv4(h, edge_index)
        h = h.tanh()  # Kích hoạt

        # Tầng 5
        h = self.conv5(h, edge_index)
        h = h.tanh()  # Kích hoạt

        # Tầng phân lớp
        out = self.classifier(h)

        return out, h

# Hàm huấn luyện
def train(data):
    model.train()
    data = data.to(device)
    optimizer.zero_grad()
    out, h = model(data.x, data.edge_index)
    loss = criterion(out[data.train_mask], data.y[data.train_mask])
    loss.backward()
    optimizer.step()

    accuracy = {}
    predicted_classes = torch.argmax(out[data.train_mask], dim=1)
    target_classes = data.y[data.train_mask]
    accuracy['train'] = torch.mean((predicted_classes == target_classes).float())

    predicted_classes = torch.argmax(out, dim=1)
    target_classes = data.y
    accuracy['val'] = torch.mean((predicted_classes == target_classes).float())

    return loss, h, accuracy

# Siêu tham số
lr = 0.01
max_epoch = 500

# Khởi tạo mô hình và chuyển sang thiết bị tính toán
model = GCN().to(device)

# Định nghĩa hàm mất mát
criterion = torch.nn.CrossEntropyLoss()

# Định nghĩa bộ tối ưu
optimizer = torch.optim.Adam(model.parameters(), lr=lr)
# Thực hiện và xem quá trình huấn luyện
for epoch in range(max_epoch):
    loss, h, accuracy = train(data)
    # Thử trực quan embedding cứ mỗi 100 epoch
    if (epoch + 1) % 100 == 0:
        visualize(h, color=data.y.cpu(), epoch=epoch+1, loss=loss, accuracy=accuracy)
        time.sleep(0.3)
````
